# Physical Properties of Nitrogen Gas

```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
from CoolProp.CoolProp import PropsSI
```

see: http://www.coolprop.org/coolprop/HighLevelAPI.html#high-level-api

```python
T = np.linspace(80,98,1000)
ρ = 100/94**3 * 1E30 * 2 * 2.3258671E-26
ρ=2.1
print(ρ)
```

```python
P = np.zeros_like(T) 
for i in range(T.size):
   P[i] = PropsSI('P','D',3.99, 'T', T[i], 'Nitrogen')
```

```python
plt.figure()
plt.plot(T,9.86923e-6*P)
plt.xlabel('T (K)')
plt.ylabel('P (atm)')
```

```python
print(9.86923e-6*P[0])
m = (9.86923e-6*(P[-1]-P[0]))/(T[-1]-T[0])
print('m =', m)
```

```python
ρρ = np.zeros_like(T) 
for i in range(T.size):
   ρρ[i] = PropsSI('D','P',101325, 'T', T[i], 'Nitrogen')
```

```python
plt.figure()
plt.plot(T,ρρ)
plt.xlabel('T (K)')
plt.ylabel(r'$\rho (kg/m3)$')
```
